Kmeans

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler 
```

```python
crime =pd.read_csv('crime_data.csv')
```

```python
crime
```

```python
crime.info()
```

```python
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
crime
```

Normalization method

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_crime_df = scaler.fit_transform(crime.iloc[:,:])
```

```python
scaled_crime_df
```

```python
kmeans = KMeans(n_clusters=3,random_state=0)
kmeans.fit(scaled_crime_df)
```

```python
kmeans.inertia_
```

```python
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i,random_state=0)
  kmeans.fit(scaled_crime_df)
  wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

Build cluster algorithm

```python
from sklearn.cluster import KMeans
clusters_new = KMeans(n_clusters=4, random_state=40)
clusters_new.fit(scaled_crime_df)
```

```python
clusters_new.labels_
```

Assign clusters to the data set

```python
crime['clusterid_new'] = clusters_new.labels_
```

Standardized values

```python
clusters_new.cluster_centers_
```

```python
crime.groupby('clusterid_new').agg(['mean'])
```

INference: In cluster 2 we can see the murder rate is low, assault rate is low,Urbanopop is low and rape rate is low and followed by cluster 0,3 and in cluster 1  murder rate is High, assault rate is High,Urbanopop is High and rape rate is High
