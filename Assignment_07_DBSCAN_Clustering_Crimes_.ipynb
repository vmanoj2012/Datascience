# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime=pd.read_csv('crime_data.csv')
crime
```

```python
crime.info()
```

```python
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
crime
```

Normalize heterogenous numerical data using standard scalar fit transform to dataset


```python
crime_norm=StandardScaler().fit_transform(crime)
crime_norm
```

 DBSCAN Clustering

```python
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(crime_norm)
```

Noisy samples are given the label -1.

```python
dbscan.labels_
```

```python
cl = pd.DataFrame(dbscan.labels_, columns = ['cluster'])
```

```python
clustered = pd.concat([crime, cl], axis = 1)
```

```python
clustered.groupby('cluster').agg(['mean'])
```

Inferences: We can tell that cluster 0 has more crime and cluster 1 has less crime rate compared to cluster 0. on com[paring both we can tell cluster 1 is safe
