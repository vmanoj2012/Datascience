# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
airline=pd.read_excel('EastWestAirlines.xlsx')
airline
```

```python
airline.info()
```

```python
airline.drop(['ID#'],axis=1,inplace=True)
airline
```

Normalize heterogenous numerical data using standard scalar fit transform to dataset


```python
airline_norm=StandardScaler().fit_transform(airline)
airline_norm
```

 DBSCAN Clustering

```python
dbscan=DBSCAN(eps=1,min_samples=11)
dbscan.fit(airline_norm)
```

Noisy samples are given the label -1.

```python
dbscan.labels_
```

```python
cl = pd.DataFrame(dbscan.labels_, columns = ['cluster'])
```

```python
clustered = pd.concat([airline, cl], axis = 1)
```

```python
clustered.groupby('cluster').agg(['mean'])
```

Inferences: From the above we can infer that cluster 1 has more balance and airlines can award the rewards to the passengers in cluster 1 and cluster 0 has less balance compared to cluster 1
