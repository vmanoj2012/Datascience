```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
forest_fires['size_category']=label_encoder.fit_transform(forest_fires['size_category'])
```

```python
forest_fires.head()
```

```python
x=forest_fires.iloc[:,0:28]
y=forest_fires.iloc[:,28]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

Neural Network - ANN

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
from keras.layers import Dropout
```

initializing ANN

```python
model=Sequential()
```

Adding input and 1st hidden layer

```python
model.add(Dense(units=10,activation='relu',kernel_initializer='he_uniform',input_dim=28))
```

Adding 2nd hidden layer

```python
model.add(Dense(units=8,activation='relu',kernel_initializer='he_uniform'))
```

Adding output layer

```python
model.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

compile the model

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

fit the model

```python
model.fit(x_train,y_train,batch_size=10,epochs=150,validation_split=0.33)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

Evaluating the model

```python
scores = model.evaluate(x_test, y_test)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
